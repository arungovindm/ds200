```python
data = pd.read_excel('./Downloads/datafile(1).xls')
```

```python
data = data.dropna(axis=0)
```

```python
data
```

```python
from matplotlib import pyplot as plt
```

```python
data = data.T
```

```python
div=[]
label = []
for i in range(1,13):
    div.append(data.iloc[2,i])
    label.append(data.columns[i])
label
```

```python
color = []
for i in range(12):
    if i%3 == 1:
        color.append(0.2)
    elif i%3==2:
        color.append(0.4)
    elif i%3 == 0:
        color.append(0.6)
```

```python
barlist = plt.barh(label,div)
for i in range(12):
    if i%3 == 1:
        barlist[i].set_color('r')
    elif i%3==2:
        barlist[i].set_color('g')
    elif i%3 == 0:
        barlist[i].set_color('b')
plt.xlabel('percetage enrolment')
plt.ylabel('classes')
plt.title('Percentage enrolment across classes in 1991-92')
plt.show()
```
